```python
import cv2
import numpy as np

# Load image
img = cv2.imread(r"C:\Users\admin\anaconda3\face.jpg")

# Resize (optional)
img = cv2.resize(img, (600, 600))

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply median blur
blur = cv2.medianBlur(gray, 5)

# Detect edges
edges = cv2.adaptiveThreshold(
    blur, 255,
    cv2.ADAPTIVE_THRESH_MEAN_C,
    cv2.THRESH_BINARY,
    9, 9
)

# Apply bilateral filter (smooth colors)
color = cv2.bilateralFilter(img, 9, 250, 250)

# Combine edges and color
cartoon = cv2.bitwise_and(color, color, mask=edges)

# Show images
cv2.imshow("Original Face", img)
cv2.imshow("Cartoon Face", cartoon)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    blur = cv2.medianBlur(gray, 5)
    
    edges = cv2.adaptiveThreshold(
        blur, 255,
        cv2.ADAPTIVE_THRESH_MEAN_C,
        cv2.THRESH_BINARY,
        9, 9
    )
    
    color = cv2.bilateralFilter(frame, 9, 250, 250)
    cartoon = cv2.bitwise_and(color, color, mask=edges)
    
    cv2.imshow("Cartoon Live", cartoon)
    
    if cv2.waitKey(1) == 27:  # Press ESC
        break

cap.release()
cv2.destroyAllWindows()
```
